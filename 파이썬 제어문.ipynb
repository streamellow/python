# while문

```python
prompt = """
    1. Add
    2. Del
    3. List
    4. Quit

    Enter number: """
```

```python
number = 0
while number != 4:
    print(prompt)
    number = int(input())
```

## break문

```python
coffee = 2
while True:
    money = int(input("돈을 넣어 주세요: "))
    if money == 300:
        print("커피를 줍니다.")
        coffee = coffee -1
    elif money > 300:
        print("거스름돈 %d를 주고 커피를 줍니다." % (money -300))
        coffee = coffee -1
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
        print("남은 커피의 양은 %d개 입니다." % coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지 합니다.")
        break
```

## continue문

```python
a = 0
while a < 10:
    a = a + 1
    if a % 2 == 0: continue
    print(a)
```

```python
k = 0
while k < 3:
    print(f"반복 중: {k}")
    k = k + 1
else:
    print("k < 3 조건이 거짓이 되어 정상 종료되었습니다.")

print("---")

l = 0
while l < 5:
    print(f"반복 중: {l}")
    if l == 2:
        break # break로 강제 종료
    l = l + 1
else:
    print("이 메시지는 출력되지 않습니다.") # break로 인해 실행 안 됨
```

### 중첩된 while문

```python
row = 1
while row <= 5: # 5개의 행을 반복 (바깥 루프)
    star = 1
    while star <= row: # 현재 'row' 개수만큼 별을 반복 (안쪽 루프)
        print("⭐️", end="") # 줄바꿈 없이 출력
        star += 1
    print() # 한 줄의 별 출력이 끝나면 줄바꿈
    row += 1
```

# for문

```python
matrix = [[1, 2], [3, 4], [5, 6]]

# 2차원 리스트를 1차원으로 평탄화
flattened = [item for row in matrix for item in row]

print(flattened) # 결과: [1, 2, 3, 4, 5, 6]
```

```python
from itertools import zip_longest

list_a = [1, 2, 3]
list_b = ['a', 'b']

# list_b의 부족한 부분을 '미정'으로 채움
for item_a, item_b in zip_longest(list_a, list_b, fillvalue='미정'):
    print(f"({item_a}, {item_b})")
```
