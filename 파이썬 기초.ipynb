```python
s11 & s22
```

```python
s11.intersection(s22)
```

```python
s11 | s22
```

```python
s11.union(s22)
```

#### 차집합

```python
s11 - s22
```

```python
s22 - s11
```

```python
s11.difference(s22)
```

```python
s22.difference(s11)
```
