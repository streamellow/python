````python
def add_mul(choice, *args):
    if choice == 'add':
        result = 0
        for i in args:
            result += i
    elif choice == 'mul':
        result = 1
        for i in args:
            result *= i
    return result
````

````python
# ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„
def rock_paper_scissors(child1, child2):
    if child1 == child2:
        return
    elif child1 == 'rock' and child2 == 'paper':
        sign = '<'
    elif child1 == 'rock' and child2 == 'scissors':
        sign = '>'
    elif child1 == 'paper' and child2 == 'rock':
        sign = '>'
    elif child1 == 'paper' and child2 == 'scissors':
        sign = '<'
    elif child1 == 'scissors' and child2 == 'rock':
        sign = '<'
    elif child1 == 'scissors' and child2 == 'paper':
        sign = '>'
    print(f'child1 : {child1} {sign} child2 : {child2}')
````
